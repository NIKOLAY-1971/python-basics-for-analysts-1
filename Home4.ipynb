```python
# Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
# Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.
```

```python
# Найдем коэффициенты корреляции, что позволит оценить зависимость стоимости домов от их характеристик.
df.drop(columns='id', inplace=True)
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix
```

```python
# Тепловая карта
plt.figure(figsize=(10,8))
sns.set(font_scale=0.7)
sns.heatmap(corr_matrix, annot=True, linewidths=0.5, cmap='coolwarm')
plt.title('Матрица корреляции');
```

```python
# Наибольшая линейная связь с ценой у следующих характеристик: sqft_living - 0.7, grade - 0.7, sqft_above - 0.6 и sqft_living15 - 0.6.
```

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

```python
# Вывод: видна прямая зависимость между размером жилой площади и ценой дома, чем больше площадь, тем выше цена.
```

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
```

```python
# Вывод: прослеживается прямая зависимость оценкой уровня качества и дизайна и ценой дома, чем больше величина уровня качества данной характеристики, тем выше цена.
```

```python
sns.jointplot(x=df['price'], y=df['sqft_above'], kind='reg');
```

```python
# Вывод: видна прямая зависимость между размером площади выше земли и ценой дома, чем больше площадь, тем выше цена.
```

```python
sns.jointplot(x=df['price'], y=df['sqft_living15'], kind='reg');
```

```python
# Вывод: видна прямая зависимость между размером жилой площади у 15 соседей и ценой дома, чем больше площадь, тем выше цена.
```
